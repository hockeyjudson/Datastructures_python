```python
import operator
class Heap:
    def __init__(self,heaplist = []):
        self._h = heaplist
        self._heap =[]
        self._op1 =  None
        self._op2 = None
        self._op3 = None
    def maxheap(self):
        self._op1 = operator.gt
        self._op2 = operator.lt
        self._op3 = operator.gt
        if self._h is not None:
            for  i in self._h:
                self.push(i)
        return self
    def minheap(self):
        self._op1 = operator.lt
        self._op2 = operator.gt
        self._op3 = operator.lt
        if self._h is not None:
            for  i in self._h:
                self.push(i)
        return self
    def push(self, element):
        self._heap.append(element)
        self._shiftup(len(self._heap) -1)
    def _shiftup(self,idx):
        parentidx = (idx-1)//2
        if parentidx< 0 :
            return
        if self._op1(self._heap[idx] , self._heap[parentidx]):
            self._swap(idx,parentidx)
            self._shiftup(parentidx)
    def _swap(self, i , j):
        self._heap[i] , self._heap[j] = self._heap[j] , self._heap[i]
    def __str__(self):
        return  ",". join(str(i) for i in self._heap)
    def __len__ (self):
        return len(self._heap)
    def _pop(self):
        self._swap(len(self._heap)-1,0)
        e = self._heap.pop()
        self._shiftdown(0)
        return e
    def _shiftdown(self,idx):
        childidx = idx*2 + 1
        if childidx >= len(self._heap):
            return
        if childidx + 1 < len(self._heap) and self._op2(self._heap[childidx] , self._heap[childidx+1]):
            childidx +=1
        if self._op3(self._heap[childidx] , self._heap[idx]):
            self._swap(childidx,idx)
            self._shiftdown(childidx)
    def sort(self):
        sorted_arr = [] 
        while len(self._heap) > 0:
            sorted_arr.append(self._pop())
        return sorted_arr
```
