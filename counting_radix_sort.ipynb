```python
#counting sort with characters
def counting_sort(ipstr):
    dct = {chr(i) :  0 for i in range(ord('a'),ord('z')+1)}
    conc_str = ""
    for i in ipstr:
        if i in dct:
            dct[i] = dct[i] +1
        else:
            conc_str = conc_str + i
    ret_str = ""
    for i in range(ord('a'),ord('z')+1):
        c = chr(i)
        ret_str = ret_str + c*(dct[c])
    return ret_str + conc_str
```

```python
#counting sort for integers
def counting_sort_int(list_int,modn):
    dt  = {i:[] for i in range(10)}
    ret_lst = []
    for i in list_int:
        j = i%modn
        #if not 
        dt[i%modn].append(i)
    for i in range(10):
        ret_lst.extend(dt[i])
    return ret_lst
```

```python
def radix_sort(iplist):
    radix = 10
    maxlength = False
    tmp,placement = -1,1
    while not maxlength:
        maxlength =True
        buckets = [list() for _ in range(radix)]
        for i in iplist:
            tmp = i // placement
            buckets[int(tmp%radix)].append(i)
            if maxlength and tmp > 0:
                maxlength =False
        a = 0
        for buck in buckets:
            for b in buck:
                iplist[a] = b
                a += 1
        placement *= radix
    return iplist
```
